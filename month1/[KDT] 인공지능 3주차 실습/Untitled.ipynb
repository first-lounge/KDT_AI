# Mission 2. 가즈아!

## Core Mission

```python
#비트코인 데이터 수집 후 이동평균법 적용
import matplotlib.dates as mdates
B_Coin = pd.read_csv("C:/Users/dlwps/AI/BitCoin.csv")
B_Coin = B_Coin.sort_values(by='Date')
B_Coin['Mov_avg'] = B_Coin['Open'].rolling(window = 5).mean()
B_Coin
```

```python
#주어진 기간의 데이터 추출
B_Coin1 = B_Coin[B_Coin["Date"].between('2016-06-01','2017-06-01')]

B_Coin1
```

```python
#해당 데이터 그래프
plt.figure(figsize=(12,7))
sns.lineplot(x=B_Coin1['Date'], y=B_Coin1["Mov_avg"], color="#f2a900")
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
plt.xticks(rotation=45)
plt.title("5-MA(Moving Average) Bit Coin Graph")
plt.xlabel('Date')
plt.ylabel('5-MA(Moving Average)')
plt.show()
```

## Extra Mission

```python
# 이더리움 데이터 수집
Eth = pd.read_csv("C:/Users/dlwps/AI/ETH_day.csv")
Eth = Eth.sort_values(by='Date')
Eth['Mov_avg'] = Eth['Open'].rolling(window = 5).mean()
Eth
```

```python
#해당 기간 추출
ETH = Eth[Eth["Date"].between('2016-06-01','2017-06-01')]
ETH
```

```python
#비트코인, 이더리움 그래프
fig = plt.figure(figsize=(12,7))
sns.lineplot(x=B_Coin1['Date'], y=B_Coin1["Mov_avg"], color="#f2a900", label="Bit_Coin")
sns.lineplot(x=ETH['Date'], y=ETH["Mov_avg"], color="#3c3c3d", label="Eth")
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
plt.xticks(rotation=45)
plt.title("5-MA(Moving Average) Bit Coin Graph")
plt.xlabel('Date')
plt.ylabel('5-MA(Moving Average)')
plt.show()
```
