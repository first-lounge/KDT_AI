## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- matplotlib
- %matplotlib inline을 통해 활성화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([2,4,1,4.5,5.1])
plt.show()
```

```python
plt.figure(figsize=(3,3)) #plotting할 도면을 선언
plt.plot([2,4,1,4.5,5.1])
plt.show()
```

## 2차함수 그래프 with plot()

```python
plt.plot([0,1,2,3,4])
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
plt.plot(x,y)
```

```python
x = np.arange(-10,10,0.01)
plt.plot(x, x**2)
plt.show()
```

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.plot(x, x**2)
plt.show()
```

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0, 25])
plt.plot(x, x**2)
plt.show()
```

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0, 25])
plt.xticks([i for i in range(-5, 6,1)])
plt.yticks([i for i in range(0,25,3)])
plt.plot(x, x**2)
plt.show()
```

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0, 25])
plt.xticks([i for i in range(-5, 6,1)])
plt.yticks([i for i in range(0,25,3)])
plt.title("y = x^2 graph")

plt.plot(x, x**2, label="trand")
plt.legend()
plt.show()
```

## II. Matplotlib Case Study

```python
x = np.arange(20)
y = np.random.randint(0,20,20)
plt.plot(x,y)
plt.axis([0,20,0,20])
plt.yticks([0,5,10,15,20])
plt.show()
```

### 산점도(scatter plot)

```python
plt.scatter(x,y)
plt.show()
```

### 박스 그림(box plot)
- 수치형 데이터에 대한 정보(Q!, Q2, Q3, min, max)

```python
plt.boxplot((x, y))
plt.title("Box plot of x, y")
plt.show()
#아래부터 최솟값, 하위 25%, 중앙값, 상위 25%, 최댓값)
```

##  막대그래프(Bar plot)

```python
plt.bar(x, y)
plt.xticks(np.arange(0, 20, 1))
plt.show()
```

```python
#히스토그램
#계급으로 나타내는 것이 특징: 범주형 데이터로 구성 후 그림을 그림
plt.hist(y, bins=np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

## 원형 그래프(Pie chart)
- 데이터 전체에 대한 부분의 비율을 부채꼴로 나타냄
- 비율 확인에 용이


```python
z = [100,200,300,400]
plt.pie(z,labels=['one', 'two', 'three', 'four'])
plt.show()
```

```python
# kdeplot()
sns.kdeplot(y, shade=True)
plt.show()
```

```python
#카운트그림 
vote_df = pd.DataFrame({"name":['Andy', 'Bob', 'Cat'],"vote":[True, True, False]})
vote_df
```

```python
vote_count = vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x=[False,True],height=vote_count['name'])
plt.show()
```

```python
sns.countplot(vote_df['vote'])
plt.show()
```
