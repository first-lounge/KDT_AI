## Core Mission

[ë‹¤ìŒ ë°ì´í„°](https://  www.kaggle.com/rishidamarla/bitcoin-prices-20112015)ê°€ ì£¼ì–´ì¡Œì„ ë•Œ 2016.6 ~ 2017.6 ê¸°ê°„ì˜ **5-MA(Moving Average) ë¹„íŠ¸ì½”ì¸ ê°€ê²© ê·¸ë˜í”„**ë¥¼ ê·¸ë ¤ì£¼ì„¸ìš”.

- ì„ ì˜ ìƒ‰ê¹”ì€ **ë…¸ë€ìƒ‰(#f2a900)** ìœ¼ë¡œ í•´ì•¼í•©ë‹ˆë‹¤.
- x-axis labelê³¼ y-axis labelì€ ì ì ˆí•œ í…ìŠ¤íŠ¸ë¡œ ì¶”ê°€ë˜ì–´ì•¼ í•©ë‹ˆë‹¤.
- ì´ë™í‰ê· ì˜ ê¸°ì¤€ì€ `Open` Columnìœ¼ë¡œ ê³„ì‚°í•©ë‹ˆë‹¤.
- ì´ì™¸ì—ë„ ê·¸ë˜í”„ë¥¼ ì•Œì•„ë³´ê¸° ì‰½ê²Œ í•˜ê¸° ìœ„í•œ ì •ë³´(Text, Facet ë“±)ì„ ì¶”ê°€í•˜ëŠ” ê²ƒë„ ì¢‹ìŠµë‹ˆë‹¤.

ğŸ’¡ ì´ë™í‰ê· (Moving Average)ë²•ì€ ì‹œê³„ì—´ ë°ì´í„°ë¥¼ í‘œí˜„í•˜ëŠ” ë°ì— ë§ì´ ì‚¬ìš©í•˜ëŠ” ë°©ë²• ì¤‘ í•˜ë‚˜ì…ë‹ˆë‹¤.

```python
# matplotlib ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import matplotlib.pyplot as plt

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
bitcoin = pd.read_csv("BitCoin.csv")

# ì£¼ì–´ì§„ ê¸°ê°„ìœ¼ë¡œ í•„í„°ë§
bitcoin_data = bitcoin[(bitcoin['Date'] >= "2016-06-01") & (bitcoin['Date'] <= "2017-06-30")]

# ë‚ ì§œ ìˆœìœ¼ë¡œ ì •ë ¬
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])
bitcoin_data = bitcoin_data.sort_values(by = 'Date')

# 5-MA ë°©ë²• ì‚¬ìš©
bitcoin_data['5MA'] = bitcoin_data['Open'].rolling(window = 5, min_periods = 1).mean()
bitcoin_data.tail(10)
```

```python
# ê·¸ë˜í”„ êµ¬í˜„

plt.figure(figsize = (16, 8))
plt.plot(bitcoin_data['Date'], bitcoin_data['5MA'], c = "#f2a900")
plt.xlabel('Date')
plt.ylabel('Price of Bitcoin')
plt.title('5-MA of Bitcoin')
plt.show()
```
